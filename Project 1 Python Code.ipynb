```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from random import randint
from sklearn.preprocessing import LabelEncoder 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
file_path = '/Users/ajayiyer/Desktop/DAT 402/Project 1/automobile.csv'

df = pd.read_csv(file_path)
df
```

```python
df.drop(df.columns[[0,1,2]], axis=1, inplace=True)
df
```

```python
df.isna().sum()
```

```python
df = df.dropna(how='any', axis=0)
df.isna().sum()
```

```python
df.rename(columns={'fuel-type': 'fuel_type', 'num-of-doors': 'num_of_doors', 'body-style': 'body', 'drive-wheels': 'drive_wheels', 'engine-location': 'engine_location', 'wheel-base': 'wheel_base', 'curb-weight': 'curb_weight', 'engine-type': 'engine', 'num-of-cylinders': 'cylinders', 'engine-size': 'engine_size', 'fuel-system': 'fuel_system', 'compression-ratio': 'compression_ratio', 'peak-rpm': 'rpm_peak', 'city-mpg': 'city_mpg', 'highway-mpg': 'highway_mpg'}, inplace=True)
df
```

```python
le = LabelEncoder()

le.fit(df.fuel_type.drop_duplicates())
df.fuel_type = le.transform(df.fuel_type)

le.fit(df.aspiration.drop_duplicates())
df.aspiration = le.transform(df.aspiration)

le.fit(df.num_of_doors.drop_duplicates())
df.num_of_doors = le.transform(df.num_of_doors)

le.fit(df.body.drop_duplicates())
df.body = le.transform(df.body)

le.fit(df.drive_wheels.drop_duplicates())
df.drive_wheels = le.transform(df.drive_wheels)

le.fit(df.engine_location.drop_duplicates())
df.engine_location = le.transform(df.engine_location)

le.fit(df.engine.drop_duplicates())
df.engine = le.transform(df.engine)

le.fit(df.cylinders.drop_duplicates())
df.cylinders = le.transform(df.cylinders)

le.fit(df.fuel_system.drop_duplicates())
df.fuel_system = le.transform(df.fuel_system)

df
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=0)
lm = LinearRegression()
lm.fit(x_train, y_train)
acc_score = lm.score(x_test, y_test)
print(acc_score)
```

```python
print("Coefficients: ", lm.coef_)
print("Intercept: ", lm.intercept_)
```
